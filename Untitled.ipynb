```python
import yfinance as yf
```

```python
import pandas as pd
```

```python
apple_minute_data = yf.download(tickers="HALKB.IS", period="2y", interval="1h", auto_adjust=True)
apple_minute_data
```

```python
# Aggregate function
ohlcv_dict = {'Open': 'first',
              'High': 'max',
              'Low': 'min',
              'Close': 'last',
              'Volume': 'sum'
             }
```

```python
# Resample data to 15 minutes data
apple_minute_data_15M = apple_minute_data.resample('4H').agg(ohlcv_dict)

# Drop the missing values
apple_minute_data_15M.dropna(inplace=True)

apple_minute_data_15M
```
